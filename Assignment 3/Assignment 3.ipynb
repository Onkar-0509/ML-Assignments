```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Admission_Predict.csv")
data.head()
```

```python
data.info()
```

```python
rating_counts = data['University Rating'].value_counts()
```

```python
plt.figure(figsize=(5, 12))
colors = sns.color_palette("pastel", len(rating_counts))
plt.pie(rating_counts.values, labels=rating_counts.index, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of University Ratings', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.figure(figsize=(10, 12))


bars = plt.bar(rating_counts.index, rating_counts.values, color=sns.color_palette("viridis", len(rating_counts)))


plt.xlabel('University Rating', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Frequency of University Ratings', fontsize=14, fontweight='bold')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}', ha='center', va='bottom', fontsize=10)
```

```python
plt.scatter(x=data['GRE Score'], y=data['TOEFL Score'])
plt.xlabel("GRE Score")
plt.ylabel("TOEFL Score")
plt.title("Comparison betweeen GRE Score and TOEFL Score")
```

```python
sns.scatterplot(x=data['GRE Score'], y=data['TOEFL Score'], hue=data['University Rating'])
plt.title("Comparison betweeen GRE Score and TOEFL Score by University Rating")
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=data[data['Research'] == 1]['SOP'], 
             y=data[data['Research'] == 1]['LOR '], 
             marker='o', linestyle='-', color='green', label='Did Research')
sns.lineplot(x=data[data['Research'] == 0]['SOP'], 
             y=data[data['Research'] == 0]['LOR '], 
             marker='s', linestyle='-', color='red', label='Did Not Do Research')
plt.xlabel("SOP (Statement of Purpose)", fontsize=12)
plt.ylabel("LOR (Letter of Recommendation Strength)", fontsize=12)
plt.title("SOP vs LOR with Research Status", fontsize=14, fontweight='bold')

plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.scatter(x=data['CGPA'], y=data['TOEFL Score'])
plt.xlabel("CGPA")
plt.ylabel("TOEFL Score")
plt.title("Comparison betweeen CGPA and TOEFL Score")
```

```python
research_yes = data[data['Research'] == 1] 
research_no = data[data['Research'] == 0]

plt.figure(figsize=(10, 6))
sns.histplot(research_yes['GRE Score'], bins=20, color='blue', label='Did Research', kde=True, stat="frequency")
sns.histplot(research_no['GRE Score'], bins=20, color='orange', label='Did Not Do Research', kde=True, stat="frequency")

plt.title('GRE Score Distribution Based on Research Participation', fontsize=16)
plt.xlabel('GRE Score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
